# Módulo 10 tarefa 2

Considere a mesma base de dados utilizada na tarefa 1 ```previsao_de_renda.csv```.

1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
````

````python

df = pd.read_csv('/content/Profissao- Cientista de Dados_M10_support material.csv')

df = df.drop(columns='Unnamed: 0').set_index(keys='index', drop=True)
df.drop_duplicates(inplace=True)
df
````

````python
plt.rc('figure', figsize=(12,12))
fig, axes = plt.subplots(2,2)

sns.pointplot(x='posse_de_imovel',
              y='renda',
              data=df,
              dodge=True,
              ax=axes[0,0])

sns.pointplot(x='posse_de_veiculo',
              y='renda',
              data=df,
              dodge=True,
              ax=axes[0,1])

sns.pointplot(x='posse_de_imovel',
              y='renda',
              hue='posse_de_veiculo',
              data=df,
              dodge=True,
              ax=axes[1,0])

sns.pointplot(x='posse_de_veiculo',
              y='renda',
              hue='posse_de_imovel',
              data=df,
              dodge=True,
              ax=axes[1,1])

plt.show()
````

Ao analisar os gráficos da primeira linha, fica evidente que ambas as variáveis categóricas, posse de veículo e posse de imóvel, exercem influência estatisticamente significativa sobre a renda. No entanto, a variável posse de veículo se destaca como mais relevante para prever a renda, conforme demonstrado pelo maior distanciamento do intervalo de confiança entre os grupos que possuem e não possuem veículo.

Para um exame mais aprofundado, os gráficos da segunda linha revelam que a variável posse de imóvel apresenta menor relevância. Isso se deve ao fato de que, entre os indivíduos que já possuem veículo, a renda se mantém relativamente estável, independentemente da posse ou não de imóvel.


**Em resumo:**

Tanto a posse de veículo quanto a posse de imóvel influenciam a renda de forma significativa.
A posse de veículo é um indicador mais preciso da renda do que a posse de imóvel.
Para indivíduos que já possuem veículo, a posse de imóvel não agrega grande valor preditivo à renda.

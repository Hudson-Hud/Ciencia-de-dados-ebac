```python
# Importando os pacotes

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Lendo o csv e transformando em um DataFrame

sinasc_RO_2019 = pd.read_csv("/content/sinasc_RO_2019.csv")
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sns.boxplot(x='IDADEMAE', data=sinasc_RO_2019)

# Mostrar o gráfico
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
sns.boxplot(x='PESO', data=sinasc_RO_2019)
```

####  3) Faça um histograma da idade da mãe

```python
# Para Criar um histograma da idade da mãe
plt.hist(x='IDADEMAE', data=sinasc_RO_2019, bins=20)

# O parâmetro bins indica o número de intervalos ou faixas que o histograma vai usar para agrupar os valores da variável

# Vamos Adicionar um título e os nomes dos eixos para ficar com uma visualização mais agradável
plt.title("Histograma da idade da mãe")
plt.xlabel("Idade da mãe")
plt.ylabel("Frequência")

# Mostrar o gráfico
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
# Vamos seguir os mesmos passos pdo item 3)

plt.hist(x='PESO', data=sinasc_RO_2019, bins=20)

plt.title("Histograma do peso do bebê")
plt.xlabel("Peso do bebê (g)")
plt.ylabel("Frequência")

plt.show()
```
